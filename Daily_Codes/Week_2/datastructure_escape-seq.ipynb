```python
line = input("Enter a string: ")
ch1 = input("Enter character 1: ")
ch2 = input("Enter character 2: ")

for ch in line:
    if ch == ch1:
        print(ch1, "character is present")
        break
else:
    print(ch1, "character is not present")

for ch in line:
    if ch == ch2:
        print(ch2, "character is present")
        break
else:
    print(ch2, "character is not present")
```

```python
line = input("Enter a string: ")
char1 = input("Enter character 1: ")
char2= input("Enter character 2: ")

found_char1 = False
found_char2 = False

for char in line:
    if char==char1:
        found_char1= True
    if char==char2:
         found_char2= True

if found_char1:
    print("char1 found in the string")
else:
    print("char1 not found")

if found_char2:
    print("char2 found in the string")
else:
    print("char2 not found")
```

```python
line = input("enter a string:")

vowel = { 'a':0 , 'e':0 , 'i':0, 'o':0 ,'u':0}
count=0

for ch in line:
    if ch=='a':
        vowel['a'] +=1
        count += 1
    if ch=='e':
        vowel['e'] +=1
        count += 1
    if ch=='i':
        vowel['i'] +=1
        count += 1
    if ch=='o':
        vowel['o'] +=1
        count += 1
    if ch=='u':
        vowel['u'] +=1
        count += 1

print("Total vowels:",count)
print("vowel counts:",vowel)
```
