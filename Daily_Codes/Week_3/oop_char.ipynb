## Inheritance

```python
class Person:
    insti = "MITU"

    def __init__ (self, age, ht):
        self.age = age
        self.ht = ht

    def show(self):
        print(f"age -> {self.age}, height -> {self.ht}, institute -> {self.insti}")
```

```python
class Ravi(Person):

    def __init__(self, age, ht, sal):
        super().__init__(age, ht)
        self.sal = sal

    def show_sal(self):
        print(f"salary is {self.sal}")
```

```python
a = Ravi(25, 5.9, 57000)
```

```python
a.show_sal()
```

```python
a.show()
```

```python
# simple menu driven with oop
# add items, view, exit (grocery list)
```

```python
class Menuapp:

    def __init__ (self):
        self.itemlist = []

    def show_menu(self):

        print("Menu\n1.add items\n2. view items\n3. exit")

    def add_items(self):
        item = input("enter the item name:")
        self.itemlist.append(item)

    def view_items(self):
        return self.itemlist
    
    def run(self):

        while True:
            
            self.show_menu()

            ch = int(input("enter the choice"))

            if ch == 1:
                self.add_items()
            elif ch == 2:
                self.view_items()
            elif ch == 3:
                break
            else:
                return 'invalid choice'
```

```python
a = Menuapp()
```

```python
a.run()
```

```python
a.itemlist
```

```python
class Person:

    def __init__ (self, age, name):

        self.__age = age
        self.__name = name

    # getter
    def get_name(self):
        return self.__name
    
    def get_age(self):
        return self.__age
    
    # setter
    def set_age(self, age):
        self.__age = age
        

    def __str__(self):
         return f"age -> {self.__age}, name -> {self.__name}"
        
```

```python
# create a menu driven app for banking simulation
# check balance, deposit money, withdraw money
#(u can add more things as u see fit to make the app better)
```

```python
class BankApp:

    def __init__(self):
        self.balance = 1000

    def show_menu(self):
        print("\n1. Check Balance")
        print("2. Deposit Money")
        print("3. Withdraw Money")
        print("4. Exit")

    def check_balance(self):
        print("Current Balance =", self.balance)

    def deposit(self):
        amount = int(input("Enter amount to deposit: "))
        self.balance += amount
        print("Amount Deposited Successfully")

    def withdraw(self):
        amount = int(input("Enter amount to withdraw: "))

        if amount <= self.balance:
            self.balance -= amount
            print("Amount Withdrawn Successfully")
        else:
            print("Insufficient Balance")

    def run(self):

        while True:

            self.show_menu()

            choice = int(input("Enter your choice: "))

            if choice == 1:
                self.check_balance()

            elif choice == 2:
                self.deposit()

            elif choice == 3:
                self.withdraw()

            elif choice == 4:
                print("Thank You")
                break

            else:
                print("Invalid Choice")
                
```

```python
a = BankApp()
```

```python
a.run
```

```python
a.run()
```
