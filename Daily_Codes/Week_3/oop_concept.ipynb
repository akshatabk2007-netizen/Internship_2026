```python
class Person:
    def __init__(self,name,age,ht):
        self.name = name
        self.__age = age     # private variable
        self.height = ht
    def show_info(self):
        return f" name -> {self.name} , age ->{self.__age} ,height ->{self.height}"
```

```python
p1 = Person('jatin',20,5.8)
p2 = Person('sachin',22,5.9)
```

```python
p1.show_info()
```

```python
"""
create a class called employee =>
attributes -> company , name ,monthly salary,job designation
here all the objects of this company will have the same common company
make salary as private variable

create 3 objects  -> e1,e2,e3 with given attributes
method -> 1. method to claculate the annual salary
          2. method to show entire data
""";
```

```python
class Employee:
    
    company='Wipro'

    def __init__(self,name,monthly_salary,job_designation):
        self.name = name
        self.monthly_salary = monthly_salary
        self.job_designation = job_designation
        
    def annual_salary(self):
        print("Annaul Salary:",self.monthly_salary * 12)
        
    def info(self):
        return f"Name=>{self.name},Company=>{self.company},Month_salary=>{self.monthly_salary},Job_Designtion=>{self.job_designation}"
    
```

```python
e1 = Employee('ABC',12000,'HR')
e2 = Employee('PQR',25000,'AI Engineer')
e3 = Employee('XYZ',40000,'Web Desginer')
```

```python
e1.info()
```

```python
e1.annual_salary()
```
