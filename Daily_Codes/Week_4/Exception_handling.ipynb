## exception / error handling

```python
# try and except
```

```python
try:
    f = open('all.txt')
except:
    print(" file does not exist")
```

```python
try:
    f =open('a1.txt')
    print(var1)
except:
    print("file does not exist")
```

```python
# we need to be specific about what kind of error we want to catch
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
except Exception:
    print("something went wrong")
```

```python
try:
    f = open('a1.txt')
    print(var1)
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
```

```python
try:
    f = open('a1.txt')
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
```

```python
# finally
```

```python
try:
    f = open('a1.txt')
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("run complete")
```

```python
# manually raise exception
```

```python
try:
    f = open('a1.txt')
    if f.name == 'a1.txt':
        raise Exception
except FileNotFoundError:
    print("file does not exist")
except Exception as e:
    print(e)
else:
    print(f.read())
    f.close()
finally:
    print("run complete")
```
