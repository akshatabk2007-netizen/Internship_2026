```python
import seaborn as sns
df = sns.load_dataset('taxis');
df
```

```python
"""
load the taxis dataset from seaborn
remove pickup and dropoff columns from the dataset
encode all the categorial columns using label encoding 
ideally create a function which encode when u pass the dataframe
""";
```

```python
df = df.drop(['pickup' , 'dropoff'], axis = 1)
```

```python
df
```

```python
def encode_data(data):
    
    le = LabelEncoder()
    
    for col in data.columns:
        if data[col].dtype == "object" or str(data[col].dtype) == "category":
            data[col] = le.fit_transform(data[col])
            
    return data

df = encode_data(df)

print(df.head())
```
