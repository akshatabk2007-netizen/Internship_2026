```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

print("Sum =", a + b)
print("Difference =", a - b)
print("Product =", a * b)

if b != 0:
    print("Quotient =", a / b)
else:
    print("Cannot divide by zero")
```

```python
c = float(input("Enter temperature in Celsius: "))

f = (c * 9/5) + 32
k = c + 273.15

print("Celsius =", c)
print("Fahrenheit =", f)
print("Kelvin =", k)
```

```python
year = int(input("Enter year: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("Leap Year")
        else:
            print("Not Leap Year")
    else:
        print("Leap Year")
else:
    print("Not Leap Year")
```

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operator: ")

if op == "+":
    print(a + b)

elif op == "-":
    print(a - b)

elif op == "*":
    print(a * b)

elif op == "/":
    if b != 0:
        print(a / b)
    else:
        print("Division by zero not possible")

else:
    print("Invalid Operator")
```

```python
n = int(input("Enter a number: "))

for i in range(1, 11):
    print(n, "x", i, "=", n * i)
```

```python
n = int(input("Enter n: "))

for num in range(2, n + 1):
    count = 0

    for i in range(1, num + 1):
        if num % i == 0:
            count += 1

    if count == 2:
        print(num)
```

```python
count = 0
sum = 0

num = int(input("Enter number (0 to stop): "))

while num != 0:
    count = count + 1
    sum = sum + num

    num = int(input("Enter number (0 to stop): "))

if count != 0:
    print("Count =", count)
    print("Sum =", sum)
    print("Average =", sum / count)
else:
    print("No numbers entered")
```

```python
n = int(input("Enter rows: "))

for i in range(1, n + 1):
    print(" " * (n - i), end="")
    print("*" * (2 * i - 1))
```
