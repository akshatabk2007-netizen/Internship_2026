```python
"""
Question 1: Create a class BankAccount with the following:

a) Attributes: account_holder, account_number, and balance.
b) Methods: deposit(amount), withdraw(amount), and get_balance().
c) withdraw() should print a message and do nothing if the amount exceeds the current balance.
d) A __str__ method that returns a readable summary of the account.

Create two BankAccount objects and perform a few deposits and withdrawals on each.
""";
```

```python
class BankAccount:

    def __init__(self, account_holder, account_number, balance):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance = self.balance + amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient Balance")
        else:
            self.balance = self.balance - amount

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Name: {self.account_holder}, Account No: {self.account_number}, Balance: {self.balance}"


acc1 = BankAccount("Akshata", 101, 5000)
acc2 = BankAccount("Rahul", 102, 3000)

acc1.deposit(1000)
acc1.withdraw(2000)

acc2.deposit(500)
acc2.withdraw(4000)

print(acc1)
print(acc2)
```

```python
"""
Question 2: Create a class Student with attributes name, roll_no, and marks (a list of 5 subject marks). Add methods:

a) calculate_percentage() — returns the average of all marks.
b) get_grade() — returns 'A' for >= 80, 'B' for >= 60, 'C' for >= 40, and 'Fail' otherwise.
c) display() — prints all details including percentage and grade.

Create a list of 3 Student objects and call display() on each.
""";
```

```python
class Student:

    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_percentage(self):
        return sum(self.marks) / len(self.marks)

    def get_grade(self):
        per = self.calculate_percentage()

        if per >= 80:
            return "A"
        elif per >= 60:
            return "B"
        elif per >= 40:
            return "C"
        else:
            return "Fail"

    def display(self):
        print("Name:", self.name)
        print("Roll No:", self.roll_no)
        print("Percentage:", self.calculate_percentage())
        print("Grade:", self.get_grade())
        print()


s1 = Student("Harshali", 1, [80, 85, 90, 75, 70])
s2 = Student("Rahul", 2, [60, 65, 70, 55, 50])
s3 = Student("Priya", 3, [35, 40, 30, 45, 38])

students = [s1, s2, s3]

for i in students:
    i.display()
```

```python
'''
Question 3: Create a class Employee with private attributes __name, __emp_id, and __salary. Add:
a) Getter and setter methods for all three attributes.
b) The salary setter should raise a ValueError if the value is below 10000.
c) A method give_raise(percent) that increases salary by the given percentage.
Demonstrate that the private attributes cannot be accessed directly, and that all access goes through the methods.
''';
```

```python
class Employee:

    def __init__(self, name, emp_id, salary):
        self.__name = name
        self.__emp_id = emp_id
        self.__salary = salary

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    def get_emp_id(self):
        return self.__emp_id

    def set_emp_id(self, emp_id):
        self.__emp_id = emp_id

    def get_salary(self):
        return self.__salary

    def set_salary(self, salary):
        if salary < 10000:
            raise ValueError("Salary must be at least 10000")
        self.__salary = salary

    def give_raise(self, percent):
        self.__salary = self.__salary + (self.__salary * percent / 100)


e1 = Employee("Akshata", 101, 20000)

print("Name:", e1.get_name())
print("Salary:", e1.get_salary())

e1.give_raise(10)

print("New Salary:", e1.get_salary())
```

```python
'''
Question 4: Build a class SecureVault that stores a private __pin and 
a private __data dictionary. It should have:
a) A method add_entry(pin, key, value) that adds to __data only if the pin matches.
b) A method get_entry(pin, key) that returns the value only if the pin matches.
c) Any wrong pin attempt should print "Access Denied" and return nothing.
''';
```

```python
class SecureVault:

    def __init__(self, pin):
        self.__pin = pin
        self.__data = {}

    def add_entry(self, pin, key, value):

        if pin == self.__pin:
            self.__data[key] = value
        else:
            print("Access Denied")

    def get_entry(self, pin, key):

        if pin == self.__pin:
            return self.__data.get(key)
        else:
            print("Access Denied")


v1 = SecureVault(1234)

v1.add_entry(1234, "Name", "Akshata")
v1.add_entry(1234, "City", "Nagpur")

print(v1.get_entry(1234, "Name"))

v1.get_entry(1111, "Name")
```

```python
'''
Question 5: Create a base class Shape with a method area()
that returns 0 and a method describe() that prints "I am a shape".
Create three subclasses:
a) Circle(radius) — overrides area() using pi * r * r.
b) Rectangle(length, width) — overrides area().
c) Triangle(base, height) — overrides area().
Each subclass should also override describe() to print something specific about itself.
Create one object of each and call both methods.
''';
```

```python
class Shape:

    def area(self):
        return 0

    def describe(self):
        print("I am a shape")


class Circle(Shape):

    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def describe(self):
        print("I am a Circle")


class Rectangle(Shape):

    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def describe(self):
        print("I am a Rectangle")


class Triangle(Shape):

    def __init__(self, base, height):
        self.base = base
        self.height = height
       
    def area(self):
        return 0.5 * self.base * self.height

    def describe(self):
        print("I am a Triangle")


c = Circle(5)
r = Rectangle(4, 3)
t = Triangle(6, 2)

c.describe()
print("Area =", c.area())

r.describe()
print("Area =", r.area())

t.describe()
print("Area =", t.area())
```

```python
'''
Question 6: Create a class Vehicle with attributes make, model, 
and year and a method start_engine() that prints "Engine started".
Create two subclasses:
a) Car(make, model, year, num_doors) — overrides start_engine()
to include the car's name in the message.
b) ElectricCar(make, model, year, battery_capacity) — overrides start_engine() 
to print "Silent start — battery at X kWh".
Use super() in both subclasses to call the parent __init__.
Create objects and test all methods.
''';
```

```python
class Vehicle:

    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started")


class Car(Vehicle):

    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def start_engine(self):
        print(self.make, self.model, "Engine started")


class ElectricCar(Vehicle):

    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity

    def start_engine(self):
        print("Silent start - battery at", self.battery_capacity, "kWh")


c1 = Car("Honda", "City", 2022, 4)
e1 = ElectricCar("Tesla", "Model3", 2024, 75)

c1.start_engine()
e1.start_engine()
```

```python
'''
Question 7: Create a base class Animal with a method sound() that prints "Some sound".
Create subclasses Dog, Cat, and Cow, each overriding sound() with their respective sounds.
Write a function make_all_sounds(animals) that takes a list of Animal 
objects and calls sound() on each.
Demonstrate that the correct method is called regardless of the object type.
''';
```

```python
class Animal:

    def sound(self):
        print("Some sound")


class Dog(Animal):

    def sound(self):
        print("Bark")


class Cat(Animal):

    def sound(self):
        print("Meow")


class Cow(Animal):

    def sound(self):
        print("Moo")


def make_all_sounds(animals):

    for i in animals:
        i.sound()


d = Dog()
c = Cat()
cw = Cow()

animals = [d, c, cw]

make_all_sounds(animals)
```

```python
'''
Question 8: Create a class Calculator with overloaded-style methods
using default arguments:
a) add(a, b, c=0) — adds two or three numbers.
b) multiply(a, b, c=1) — multiplies two or three numbers.
c) Also implement __add__ on a class Vector(x, y) so that
two Vector objects can be added with the + operator. Print the resulting vector
''';
```

```python
class Calculator:

    def add(self, a, b, c=0):
        return a + b + c

    def multiply(self, a, b, c=1):
        return a * b * c


class Vector:

    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def display(self):
        print("(", self.x, ",", self.y, ")")


c = Calculator()

print(c.add(10, 20))
print(c.add(10, 20, 30))

print(c.multiply(2, 3))
print(c.multiply(2, 3, 4))

v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2

v3.display()
```

```python
'''
Question 10: Build a menu-driven Student Record Manager using OOP and file handling. The app should have a class StudentManager with the following menu options:
Add Student — takes name, roll number, and marks for 3 subjects.
View All — displays all stored students with their average marks.
Search Student — finds a student by roll number.
Delete Student — removes a student by roll number.
Save to File — writes all records to students_data.txt.
Load from File — reads records back from students_data.txt on startup.
''';
```

```python
class StudentManager:

    def __init__(self):
        self.students = []

    def add_student(self):
        name = input("Enter Name: ")
        roll = input("Enter Roll No: ")

        m1 = int(input("Enter Marks 1: "))
        m2 = int(input("Enter Marks 2: "))
        m3 = int(input("Enter Marks 3: "))

        self.students.append([name, roll, m1, m2, m3])

    def view_all(self):

        for s in self.students:
            avg = (s[2] + s[3] + s[4]) / 3

            print("Name:", s[0])
            print("Roll No:", s[1])
            print("Average:", avg)
            print()

    def search_student(self):

        roll = input("Enter Roll No: ")

        for s in self.students:
            if s[1] == roll:
                print(s)

    def delete_student(self):

        roll = input("Enter Roll No: ")

        for s in self.students:
            if s[1] == roll:
                self.students.remove(s)
                print("Student Deleted")

    def save_file(self):

        f = open("students_data.txt", "w")

        for s in self.students:
            f.write(str(s) + "\n")

        f.close()

    def run(self):

        while True:

            print("\n1.Add Student")
            print("2.View All")
            print("3.Search Student")
            print("4.Delete Student")
            print("5.Save To File")
            print("6.Exit")

            ch = int(input("Enter Choice: "))

            if ch == 1:
                self.add_student()

            elif ch == 2:
                self.view_all()

            elif ch == 3:
                self.search_student()

            elif ch == 4:
                self.delete_student()

            elif ch == 5:
                self.save_file()

            elif ch == 6:
                43
                break

            else:
                print("Invalid Choice")


obj = StudentManager()
obj.run()
```
