```python
'''
Write a function safe_divide(a, b) that divides two numbers and
handles ZeroDivisionError and TypeError separately, printing an appropriate message
for each. Use a finally block to print 'Operation attempted.' regardless of the outcome. 
Test it with valid numbers, division by zero, and a string input
''';
```

```python
def safe_divide(a, b):

    try:
        print("Result =", a / b)

    except ZeroDivisionError:
        print("Cannot divide by zero")

    except TypeError:
        print("Invalid data type")

    finally:
        print("Operation attempted.")


safe_divide(10, 2)
safe_divide(10, 0)
safe_divide(10, "a")
```

```python
'''
Write a function validate_age(age) that raises a custom exception InvalidAgeError
(a class you define, inheriting from Exception) if age is negative or greater than 120. 
Catch this exception in a calling block and print a user-friendly message.
Test it with at least 3 different values.

''';
```

```python
class InvalidAgeError(Exception):
    pass


def validate_age(age):

    if age < 0 or age > 120:
        raise InvalidAgeError("Invalid Age")

    print("Valid Age =", age)


ages = [20, -5, 150]

for age in ages:

    try:
        validate_age(age)

    except InvalidAgeError:
        print("Please enter a valid age")
```

```python
'''
Write a program that repeatedly asks the user (use a loop with at
least 3 simulated inputs in a list, e.g. user_inputs = ['10', 'abc', '0', '5'])
to enter a number and divide 100 by it.
Handle the following in a single try block using multiple except clauses:
a) ValueError if the input is not a number.
b) ZeroDivisionError if the input is zero.
c) A generic Exception as a catch-all for anything else.
Print the result if successful, otherwise print which error occurred.
''';
```

```python
user_inputs = ['10', 'abc', '0', '5']

for i in user_inputs:

    try:
        num = int(i)

        result = 100 / num

        print("Result =", result)

    except ValueError:
        print("ValueError: Not a number")

    except ZeroDivisionError:
        print("ZeroDivisionError: Cannot divide by zero")

    except Exception:
        print("Some other error occurred")
```

```python
'''
Create a 1D NumPy array of integers from 1 to 20. Perform the following:
a) Reshape it into a 4x5 2D array.
b) Print the sum of each row and each column.
c) Extract all elements greater than 10 using boolean indexing.
d) Replace all even numbers in the array with -1.
''';
```

```python
import numpy as np

arr = np.arange(1, 21)

print("Original Array:")
print(arr)

arr2 = arr.reshape(4, 5)

print("4x5 Array:")
print(arr2)

print("Row Sum:")
print(arr2.sum(axis=1))

print("Column Sum:")
print(arr2.sum(axis=0))

print("Elements Greater Than 10:")
print(arr[arr > 10])

arr[arr % 2 == 0] = -1

print("After Replacing Even Numbers:")
print(arr)
```

```python
'''
Create two NumPy arrays A and B as shown below and perform:
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
a) Element-wise addition, subtraction, and multiplication.
b) Matrix multiplication using np.dot() or the @ operator.
c) Transpose of A.
d) The determinant and inverse of A using np.linalg
''';
```

```python
import numpy as np

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Addition:")
print(A + B)

print("Subtraction:")
print(A - B)

print("Multiplication:")
print(A * B)

print("Matrix Multiplication:")
print(np.dot(A, B))

print("Transpose of A:")        # transpose means rows beccomes columns
print(A.T)

print("Determinant of A:")
print(np.linalg.det(A))

print("Inverse of A:")
print(np.linalg.inv(A))
```

```python
'''
Generate a NumPy array of 15 random integers between 1 and 100 using np.random.randint()
(set a seed of 42 for reproducibility). Then:
a) Find the mean, median, standard deviation, min, and max.
b) Sort the array in descending order.
c) Find the indices of the 3 largest values using np.argsort().
d) Normalize the array so all values lie between 0 and 1.
''';
```

```python
import numpy as np

np.random.seed(42)        # Set a fixed starting point for random numbers

arr = np.random.randint(1, 101, 15)

print("Array:")
print(arr)

print("Mean =", np.mean(arr))
print("Median =", np.median(arr))
print("Standard Deviation =", np.std(arr))
print("Minimum =", np.min(arr))
print("Maximum =", np.max(arr))

print("Sorted Array Descending:")
print(np.sort(arr)[::-1])

print("Indices of 3 Largest Values:")        # Returns positions indices after sorting.
print(np.argsort(arr)[-3:])

# Normalizing means converting all values between 0 and 1.
# (value - minimum) / (maximum - minimum)

normalized = (arr - np.min(arr)) / (np.max(arr) - np.min(arr))

print("Normalized Array:")
print(normalized)
```

```python
'''
Create a Pandas DataFrame from the dictionary below and perform the following:
data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67],
}
a) Display the first 3 rows and the data types of each column.
b) Add a new column Grade based on Score (>=90 'A', >=75 'B', else 'C').
c) Sort the DataFrame by Score in descending order.
d) Filter rows where Department is 'AI'.
''';
```

```python
import pandas as pd

data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)

print("First 3 Rows:")
print(df.head(3))

print("\nData Types:")
print(df.dtypes)

grade = []

for i in df["Score"]:
    if i >= 90:
        grade.append("A")
    elif i >= 75:
        grade.append("B")
    else:
        grade.append("C")

df["Grade"] = grade

print("\nDataFrame with Grade:")
print(df)

print("\nSorted by Score:")
print(df.sort_values("Score", ascending=False))

print("\nDepartment = AI")
print(df[df["Department"] == "AI"])
```

```python
'''
Using the same DataFrame from Q7:
a) Group by Department and find the average Score for each.
b) Find the row with the highest Score using idxmax().
c) Rename the column Score to Marks.
d) Drop the Age column and display the resulting DataFrame.
''';
```

```python
import pandas as pd

data = {
    "Name": ["Aarav", "Priya", "Karan", "Meera", "Sahil"],
    "Age": [23, 21, 25, 22, 24],
    "Department": ["AI", "Data Science", "AI", "Web Dev", "Data Science"],
    "Score": [88, 92, 79, 95, 67]
}

df = pd.DataFrame(data)

print("Average Score by Department:")                   # Average score by department
print(df.groupby("Department")["Score"].mean())


index = df["Score"].idxmax()                 # Row with highest score

print("\nStudent with Highest Score:")

print(df.loc[index])

df = df.rename(columns={"Score": "Marks"})        #  Rename Score to Marks

print("\nAfter Renaming:")
print(df)

df = df.drop("Age", axis=1)                                # Drop Age column

print("\n After Dropping Age:")
print(df)
```

```python
'''
Create a DataFrame with some missing values as shown below and handle them:
import numpy as np

data = {
    "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
    "Price": [10, np.nan, 5, 8, np.nan],
    "Stock": [100, 50, np.nan, 30, 20]
}
a) Check the number of missing values in each column using isnull().sum().
b) Fill missing Price values with the column mean.
c) Drop rows where Stock is still missing.
d) Save the cleaned DataFrame to a CSV file named cleaned_products.csv without the index.
''';
```

```python
import pandas as pd
import numpy as np

data = {
    "Product": ["Pen", "Notebook", "Eraser", "Sharpener", "Ruler"],
    "Price": [10, np.nan, 5, 8, np.nan],
    "Stock": [100, 50, np.nan, 30, 20]
}

df = pd.DataFrame(data)

print("Missing Values:")
print(df.isnull().sum())

mean_price = df["Price"].mean()
df["Price"] = df["Price"].fillna(mean_price)

df = df.dropna(subset=["Stock"])

print("\nCleaned DataFrame:")
print(df)

df.to_csv("cleaned_products.csv", index=False)

print("\nFile Saved Successfully")
```

```python
'''
Create a DataFrame with sales data for 6 months and generate visualizations
using Matplotlib:
a) A line chart showing sales trend over months.
b) A bar chart comparing monthly sales.
c) A pie chart showing the percentage contribution of each month to total sales.
d) Add appropriate titles, axis labels, and legends where applicable.
''';
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "Sales": [100, 150, 120, 180, 200, 170]
}

df = pd.DataFrame(data)

plt.plot(df["Month"], df["Sales"], marker="o", label="Sales")
plt.title("Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()                 # Legend -> legend tells us what each line, bar, or color
                                          #in a graph represents
plt.show()

plt.bar(df["Month"], df["Sales"])
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

plt.pie(df["Sales"], labels=df["Month"])
plt.title("Sales Contribution by Month")
plt.show()
```
