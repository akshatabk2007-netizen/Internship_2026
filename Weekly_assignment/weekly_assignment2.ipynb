```python
def analyze_sentence(sentence):

    words = sentence.lower().split()

    word_count = len(words)

    char_count = len(sentence.replace(" ", ""))

    unique_words = len(set(words))

    max_count = 0
    most_frequent_word = ""

    for word in words:
        count = words.count(word)

        if count > max_count:
            max_count = count
            most_frequent_word = word

    return {
        "word_count": word_count,
        "char_count": char_count,
        "unique_words": unique_words,
        "most_frequent_word": most_frequent_word
    }
```

```python
text = "A man a plan a canal Panama"

# a) Reverse the string
reverse = ""

for ch in text:
    reverse = ch + reverse

print("Reversed String:", reverse)

# b) Palindrome check
new_text = text.lower().replace(" ", "")

if new_text == new_text[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")

# c) Replace vowels with *
new_string = ""

for ch in text:
    if ch.lower() in "aeiou":
        new_string += "*"
    else:
        new_string += ch

print("After Replacing Vowels:", new_string)

# d) Count vowels
a = e = i = o = u = 0

for ch in text.lower():
    if ch == 'a':
        a += 1
    elif ch == 'e':
        e += 1
    elif ch == 'i':
        i += 1
    elif ch == 'o':
        o += 1
    elif ch == 'u':
        u += 1

print("a =", a)
print("e =", e)
print("i =", i)
print("o =", o)
print("u =", u)
```

```python
def format_report(data):

    print("Name    Score   Grade")

    for name, score, grade in data:
        print(f"{name:<15}{score:>6}{grade:^8}")

data = [ ("Alice", 92, "A"),("Bob", 78, "B+"),("Charlie", 55, "C"),("Diana", 100, "A+")]

format_report(data)
```

```python
def list_ops(nums):

    # a) Remove duplicates
    unique = []

    for i in nums:
        if i not in unique:
            unique.append(i)

    print("Without Duplicates:", unique)

    # b) Even index elements squared
    square_list = []

    for i in range(0, len(nums), 2):
        square_list.append(nums[i] ** 2)

    print("Even Index Squared:", square_list)

    # c) Flatten nested list
    nested = [[1, 2, 3], [4, 5], [6, 7, 8, 9]]

    flat = [item for sublist in nested for item in sublist]

    print("Flattened List:", flat)

nums = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]

list_ops(nums)    
```

```python
employee = ("E042", "Riya Sharma", "Data Science", 72000, "Pune")

# a) Unpacking

emp_id, name, department, salary, city = employee

print(emp_id)
print(name)
print(department)
print(salary)
print(city)

# b) * unpacking

first, *middle, last = employee

print("First =", first)
print("Middle =", middle)
print("Last =", last)

# c) Update salary

def update_salary(emp_tuple, new_salary):

    return (
        emp_tuple[0],
        emp_tuple[1],
        emp_tuple[2],
        new_salary,
        emp_tuple[4]
    )

new_employee = update_salary(employee, 80000)

print(new_employee)
```

```python
batch_a = {"Aarav", "Priya", "Karan", "Meera", "Sahil"}

batch_b = {"Karan", "Meera", "Tanvi", "Rohan"}

batch_c = {"Priya", "Aarav"}

# a)

print("Common Students:")
print(batch_a & batch_b)

# b)

print("Only in batch_a:")
print(batch_a - batch_b)

# c)

print("All Unique Students:")
print(batch_a | batch_b)

# d)

print("Is batch_c subset of batch_a?")
print(batch_c.issubset(batch_a))
```

```python
def inventory_manager(records):

    inventory = {}

    for product, category, quantity, price in records:

        if category not in inventory:

            inventory[category] = {
                "total_items": 0,
                "total_value": 0,
                "products": []
            }

        inventory[category]["total_items"] += quantity
        inventory[category]["total_value"] += quantity * price
        inventory[category]["products"].append(product)

    return inventory


records = [
    ("Laptop", "Electronics", 10, 55000),
    ("Mouse", "Electronics", 50, 499),
    ("Desk", "Furniture", 8, 12000),
    ("Chair", "Furniture", 15, 4500),
    ("Pen", "Stationery", 200, 20)
]

print(inventory_manager(records))
```

```python
def power_all(*args, exp=2):

    result = []

    for i in args:
        result.append(i ** exp)

    return result
```
